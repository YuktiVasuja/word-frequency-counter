```python
# Word Frequency Analysis Mini Project

import matplotlib.pyplot as plt
from collections import Counter
import re
import pandas as pd

# Sample text (replace with any paragraph)
text = """
Machine learning is a field of artificial intelligence.
Machine learning allows systems to learn automatically.
Research projects often use machine learning for predictions.
"""

# Preprocess text (lowercase + keep only words)
words = re.findall(r"\b[a-zA-Z]+\b", text.lower())

# Count word frequency
word_counts = Counter(words)

# Convert to DataFrame for better handling
df = pd.DataFrame(word_counts.items(), columns=['Word', 'Count']).sort_values('Count', ascending=False)

# Print results
print("Word Frequency Table:\n")
print(df)

# Plot bar chart
plt.figure(figsize=(8,4))
plt.bar(df['Word'], df['Count'])
plt.title("Word Frequency")
plt.xlabel("Word")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```
